```python
import cv2
from matplotlib import pyplot as plt

#Opening the image from file
image = cv2.imread("/content/rose.png")

#Convert the image to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Applying mean filter using cv2.blur
mean_smoothed_image = cv2.blur(image_rgb, (10, 10))  # Adjust the kernel size

#Resizing the images
width, height = 300, 200  # Adjust the width and height
resized_image = cv2.resize(image_rgb, (width, height))
resized_mean_smoothed_image = cv2.resize(mean_smoothed_image, (width, height))

#Displaying the original and mean-smoothed images side by side
plt.imshow(resized_image)
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.imshow(resized_mean_smoothed_image)
plt.title("Mean Filtered Image")
plt.axis('off')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

#Opening the image
image = cv2.imread("/content/rose.png")

#Convert the image to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Applying Median filter
median_smoothed_image = cv2.medianBlur(image_rgb, 9)  # Adjust the kernel size

# Resize the images
width, height = 300, 200  # Adjust the width and height
resized_image = cv2.resize(image_rgb, (width, height))
resized_median_smoothed_image = cv2.resize(median_smoothed_image, (width, height))

# Displaying the original and median-smoothed images side by side
plt.imshow(resized_image)
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.imshow(resized_median_smoothed_image)
plt.title("Median Filtered Image")
plt.axis('off')
plt.show()
```

```python
#code begins by importing the necessary libraries
#cv2 is the OpenCV library used for image processing
#matplotlib.pyplot is used for displaying images
import cv2
from matplotlib import pyplot as plt

#Opening the image
image = cv2.imread("/content/rose.png")

# Convert the image to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Applying Gaussian filter
gaussian_smoothed_image = cv2.GaussianBlur(image_rgb, (9, 9), 9)  # Adjust the kernel size as needed

# Resize the images
width, height = 300, 200  # Adjust the width and height as needed
resized_image = cv2.resize(image_rgb, (width, height))
resized_gaussian_smoothed_image = cv2.resize(gaussian_smoothed_image, (width, height))

# Displaying the original and Gaussian-smoothed images side by side
#This code block displays the resized original image using Matplotlib. The title is
#gset to "Original Image," and the axis is turned off for a cleaner display.
plt.imshow(resized_image)
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.imshow(resized_gaussian_smoothed_image)
plt.title("Gaussian Filtered Image")
plt.axis('off')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

#Opening the image file named rose.png
image = cv2.imread("/content/rose.png")

#Convert the image to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Applying bilateral filter
#d: Diameter of each pixel neighborhood
#sigmaColor: Controls how different colors are considered
#sigmaSpace: Controls how far away pixels influence each other in the spatial domain
bilateral_smoothed_image = cv2.bilateralFilter(image_rgb, d=30, sigmaColor=75, sigmaSpace=100)

#Resizing the images
width, height = 300, 200  # Adjust the width and height as needed unit pixels
resized_image = cv2.resize(image_rgb, (width, height))
resized_bilateral_smoothed_image = cv2.resize(bilateral_smoothed_image, (width, height))

#Displaying the original and bilateral-smoothed images side by side
plt.imshow(resized_image)
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.imshow(resized_bilateral_smoothed_image)
plt.title("Bilateral Filtered Image")
plt.axis('off')
plt.show()
```
